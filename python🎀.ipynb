```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_excel("C:\\Users\\TechnoHouse\\Downloads\\Call Center Data 2.xlsx", engine="openpyxl")


print(df.head())
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x=df["Language"], order=df["Language"].value_counts().index, palette="viridis")
plt.title("Language Distribution")
plt.show()

```

```python
plt.figure(figsize=(12,6))
df.groupby("Month")["Case ID"].count().plot(kind="line", marker="o", color="b")
plt.title("Number of Cases per Month")
plt.xlabel("Month")
plt.ylabel("Number of Cases")
plt.show() # نضيف اكتر من شهر
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.histplot(df["Total Work Minutes"], bins=30, kde=True, color="hotpink")
plt.title("Work Time Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df["Total Work Minutes"], y=df["Total Wrap Minutes"], alpha=0.6)
plt.title("Work vs. Wrap Time Relationship")
plt.show()
```

```python
plt.figure(figsize=(12,6))
df.groupby("Agent")["AHT"].mean().sort_values().tail(10).plot(kind="barh", color="skyblue")
plt.title("Top 10 Agents by Average AHT")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["Chat Engaged"], bins=20, kde=True, color="hotpink")
plt.title("Chat Engagement Distribution")
plt.show() #*
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y=df["Consult Outcome"], order=df["Consult Outcome"].value_counts().index, palette="coolwarm")
plt.title("Most Common Consultation Outcomes")
plt.show()
```

```python
plt.figure(figsize=(12,6))
df[["1 Star Count", "2 Star Count", "3 Star Count", "4 Star Count", "5 Star Count"]].sum().plot(kind="bar", color=["red", "orange", "yellow", "green", "blue"])
plt.title("Customer Ratings Distribution")
plt.show() # تقييمات العمله حسب النجوم
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["Helped Resolve Surveys"], bins=20, kde=True, color="hotpink")

plt.title("Helped Resolve Surveys Distribution")
plt.show()
#عدد الاستبيانات التي ساعدت في حل المشاكل 
```

```python
plt.figure(figsize=(12,6))
df.groupby("Agent")["CSAT Star Average"].mean().sort_values().plot(kind="bar", color="coral")
plt.title("CSAT Average per Agent")
plt.show()
#متوسط تقييم العملاء (CSAT Star Average) لكل وكيل (Agent).
```

```python
plt.figure(figsize=(12,6))
df.groupby("Language")["AHT"].mean().sort_values().plot(kind="bar", color="skyblue")
plt.title("Average AHT per Language")
plt.show()
#متوسط زمن التعامل مع العميل (AHT) لكل لغة (Language).
```

```python
plt.figure(figsize=(12,6))
top_10_agents = df.groupby("Agent")["AHT"].mean().nlargest(10).index  

df_top_10 = df[df["Agent"].isin(top_10_agents)]  

df_top_10.groupby(["Month", "Agent"])["AHT"].mean().unstack().plot(
    kind="line", marker="o", figsize=(12,6)
)

plt.title("Top 10 Agents: AHT Trend Over Time")
plt.xlabel("Month")
plt.ylabel("Average AHT")
plt.legend(title="Agent", bbox_to_anchor=(1,1))
plt.show()
# معدل التحسين و التراجع لكل AHT لكل وكيل فى الشهر
```

```python
df["Lost Time"] = df["Total Work Minutes"] - df["Total Wrap Minutes"]
plt.figure(figsize=(12,6))
df.groupby("Agent")["Lost Time"].sum().sort_values().plot(kind="barh", color="purple")
plt.title("Total Lost Time per Agent")
plt.show()#تحليل الوقت الضائع لكل وكيل
```

```python
df["Weekday"] = pd.to_datetime(df["Date"]).dt.day_name()
plt.figure(figsize=(10,5))
sns.countplot(x=df["Weekday"], order=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"], palette="Set2")
plt.title("Number of Cases by Weekday")
plt.show()#أي يوم في الأسبوع يكون فيه الضغط أعلى؟
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df["Language"], y=df["Helped Resolve Surveys"], palette="coolwarm")
plt.title("Helped Resolve Surveys by Language")
plt.xticks(rotation=45)
plt.show()#تحليل اللغة وتأثيرها على معدل حل المشكلات
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    x=df["Language"], 
    y=df["Total Work Minutes"],  
    palette="coolwarm",
    width=0.5,  
    linewidth=1  
)

plt.title("Total Work Minutes by Language", fontsize=12, fontweight="bold")  
plt.xticks(rotation=45, fontsize=10)  
plt.yticks(fontsize=10)  
plt.grid(axis="y", linestyle="--", alpha=0.5)  
plt.show()
```

```python
plt.figure(figsize=(12,6))
df.groupby("Agent")[["1 Star Count", "2 Star Count"]].sum().sum(axis=1).sort_values().tail(10).plot(kind="bar", color="crimson")
plt.title("Top 10 Agents with Highest Negative Ratings")
plt.xlabel("Agent")
plt.ylabel("Total Negative Ratings")
plt.show()
#تحليل تأثير الاستطلاعات السلبية على الوكلاء
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")  
df = df.dropna(subset=["Date"])  

df["Month"] = df["Date"].dt.month_name()  
df["Weekday"] = df["Date"].dt.day_name()  


week_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]


plt.figure(figsize=(12,6))
sns.countplot(x=df["Weekday"], hue=df["Month"], order=week_order, palette="coolwarm")

plt.title("Number of Cases by Weekday (Grouped by Month)")
plt.xlabel("Weekday")
plt.ylabel("Count")
plt.legend(title="Month", bbox_to_anchor=(1,1))  
plt.show()#تحليل توزع الحالات خلال أيام الأسبوع
```

```python
df["Survey Success Rate"] = df["Helped Resolve Surveys"] / df["Survey Count"]
plt.figure(figsize=(12,6))
sns.barplot(x=df["Agent"], y=df["Survey Success Rate"], palette="Blues")
plt.xticks(rotation=90)
plt.title("Survey Success Rate per Agent")
plt.show()#تحليل نسبة الاستطلاعات الناجحة إلى إجمالي الشات
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df["Total Work Minutes"], y=df["CSAT Star Average"], alpha=0.5, color="royalblue")
plt.title("Work Time vs CSAT Score")
plt.xlabel("CSAT Star Average")
plt.ylabel("Total Work Minutes")
plt.show ()
```

```python
plt.figure(figsize=(12,6))
df["Agent"].value_counts().head(10).plot(kind="bar", color="teal")
plt.title("Top 10 Agents by Number of Cases Handled")
plt.xlabel("Agent")
plt.ylabel("Number of Cases")
plt.xticks(rotation=45)
plt.show()#تحليل الوكلاء الأكثر إنتاجية بناءً على عدد الحالات
```

```python
plt.figure(figsize=(15,20))

df.groupby("Agent")[["Total Work Minutes", "Total Wrap Minutes"]].sum()\
  .nlargest(10, "Total Work Minutes")\
  .plot(kind="bar", stacked=True, color=["steelblue", "salmon"])

plt.title("Top 10 Agents: Work Time vs Wrap Time")
plt.ylabel("Minutes")
plt.xticks(rotation=90)
plt.legend(title="Time Type")
plt.show()
#تحليل توزيع زمن العمل مقابل زمن الـ Wrap Time لكل وكيل
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=df["Language"], y=df["CSAT Star Average"], palette="coolwarm")
plt.title("CSAT Star Average per Language")
plt.xticks(rotation=45)
plt.show()# تحليل أداء اللغات المختلفة بناءً على الـ CSAT
```

```python
df.groupby("Date")[["1 Star Count", "5 Star Count"]].sum().plot(kind="line", marker="o", figsize=(12,6))
plt.title("Monthly Trend of 1-Star and 5-Star Ratings")
plt.xlabel("Date")
plt.ylabel("Count")
plt.legend(["1 Star", "5 Star"])
plt.show() #تحليل التوزيع الشهري للاستطلاعات الإيجابية والسلبية
```

```python

df["Resolution Rate"] = df["Helped Resolve Surveys"] / df["Survey Count"]
df["Resolution Rate"] = df["Resolution Rate"].fillna(0)  
top_10_agents = df.groupby("Agent")["Resolution Rate"].mean().sort_values(ascending=False).head(10)
print(top_10_agents)
plt.figure(figsize=(12,6))
top_10_agents.plot(kind="barh", color="deepskyblue")  
plt.title("Top 10 Agents by Resolution Rate")
plt.xlabel("Resolution Rate")
plt.ylabel("Agent")
plt.gca().invert_yaxis()  
plt.show()

# تحليل توزيع معدلات حل المشاكل بالنسبة لكل وكيل
```

```python
plt.figure(figsize=(12,6))

top_10_agents = df.groupby("Agent")["AHT"].mean().nlargest(10).index  
df_top_10 = df[df["Agent"].isin(top_10_agents)]  

df_top_10.groupby(["Month", "Agent"])["AHT"].mean().unstack().T.plot(
    kind="barh", figsize=(12,6), colormap="Blues" 
)

plt.title("Top 10 Agents AHT Performance Over Time")
plt.xlabel("Month")
plt.ylabel("Average AHT")
plt.legend(title="Agent", bbox_to_anchor=(1,1))
plt.show()

# تحليل مدى استقرار أداء كل وكيل على مدار الأشهر
```
